#Locistik Regrasyon

```python
import numpy as np
import pandas as pd
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

#Veri Seti Oluşturma

```python
X, y = make_classification(
    n_samples=1000,
    n_features=10,
    n_informative=2,
    n_classes=2,
    random_state=42
)
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lr_model  = LogisticRegression()
lr_model.fit(x_train, y_train)
```

```python
y_pred = lr_model.predict(x_test)
basarı = accuracy_score(y_test, y_pred)
print(f"Modelin Başarı Oranı: {basarı}")
```

#KNN

```python
from sklearn.neighbors import KNeighborsClassifier

knnn_model = KNeighborsClassifier(n_neighbors=5)
knnn_model.fit(x_train, y_train)
```

```python
y_pred = knnn_model.predict(x_test)
basarı = accuracy_score(y_test, y_pred)
print(f"Modelin Başarı Oranı: {basarı}")
```

#SVR

```python
from sklearn.svm import SVC

svm_model = SVC()
svm_model.fit(x_train, y_train)
```

```python
y_pred = svm_model.predict(x_test)
basarı = accuracy_score(y_test, y_pred)
print(f"Modelin Başarı Oranı: {basarı}")
```

```python
veri.columns = ["Yas", "Amaliyat Yılı", "Bulgular", "canli"]
print(veri.head(10))
```

```python
veri.loc[veri['canli'] == 2, 'canli'] = 0
print(veri.head(10))
```

```python
print(veri.dtypes)


```

```python
from sklearn.preprocessing import MinMaxScaler

X = veri.drop('canli', axis=1)
y = veri['canli']

ölceklendirme = MinMaxScaler()
x = ölceklendirme.fit_transform(X)
x = pd.DataFrame(X, columns = ["Yas", "Amaliyat Yılı", "Bulgular"])
print(x.head(10))
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=42)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(128, input_shape = (x_train.shape[1],),activation = 'relu'))# Giriş katmanı
model.add(Dense(64, activation = 'relu'))#Gizli katman 1
model.add(Dense(32, activation = 'relu'))#Gizli katman 2
model.add(Dense(16, activation = 'relu'))#Çıkış katmanı3
model.add(Dense(1, activation = 'sigmoid'))#Çıkış katmanı
model.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])

```

```python
model.fit(x_train, y_train, epochs=10000, verbose=0)
```

```python
kayıp_basari = model.evaluate(x_test, y_test)
print(f"Kayıp: {kayıp_basari[0]}, Başarı: {kayıp_basari[1]}")
```
