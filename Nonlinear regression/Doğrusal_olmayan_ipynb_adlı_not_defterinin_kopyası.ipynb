Kütüphaneler


```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
import numpy as np
import pandas as pd
```

```python
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['Target'] = iris.target
data.head(10)


```

```python
X = iris.data
y = iris.target

x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

Model

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
y_pred_knn = knn.predict(x_test)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred_knn))
```

SVR

```python
from sklearn.svm import SVR
svr = SVR(kernel='linear', epsilon=0.01)
svr.fit(x_train, y_train)
y_pred_svr = svr.predict(x_test)
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test, y_pred_svr))
```

# Yapay Sinir Ağı

```python
import tensorflow as tf
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.preprocessing import StandardScaler

giris_verileri = np.array([[0], [1], [2], [3], [4], [5]])
cikis_verileri = np.array([0, 1, 4, 9, 16, 25])


model = Sequential([
    Dense(units=1, input_shape=(1,))
])
```

Optimizasyon

```python
model.compile(optimizer= 'sgd', loss = 'mean_squared_error')
```

Eğitim

```python
model.fit(giris_verileri, cikis_verileri, epochs=1000)

```

```python
print("Tahminler:")
tahminler = model.predict(np.array([[4],[5],[6]]))
print(tahminler)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

model_lineer=LinearRegression()
model_lineer.fit(giris_verileri,cikis_verileri)

print(model_lineer.predict(np.array([[6],[7],[8]])))

mse=mean_squared_error(cikis_verileri,model_lineer.predict(giris_verileri))
print("Hata puanı:",mse)

```

```python
from sklearn.tree import DecisionTreeRegressor, plot_tree
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error

x = iris.data
y = iris.target
model = DecisionTreeRegressor()
model.fit(x, y)

```
